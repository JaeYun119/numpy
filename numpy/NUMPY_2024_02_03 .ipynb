```python
import time

num = range(10000)
result = []
start = time.time()
for i in num:
  i += 3
  result.append(i)
end = time.time()
sec = (end - start) * 1000
print(f"{sec: .4f} sec")
print(result[-5:])
```

```python
import time
import numpy as np

num = np.arange(10000)
start = time.time()
result = num + 3
end = time.time()
sec = (end - start) * 1000
print(f"{sec: .5f} sec")
print(result[-5:])
```
